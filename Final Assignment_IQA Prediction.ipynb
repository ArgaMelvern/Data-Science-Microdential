```python
def show_boxplot(x):
  red_circle = dict(markerfacecolor='blue', marker='o', markeredgecolor='white')

  fig, axs = plt.subplots(1, len(x.columns), figsize=(20,5))

  for i, ax in enumerate(axs.flat):
      ax.boxplot(x.iloc[:,i], flierprops=red_circle)
      ax.set_title(x.columns[i], fontsize=10, fontweight='bold')
      ax.tick_params(axis='y', labelsize=8)
      
  plt.tight_layout()
```

```python
from sklearn.model_selection import train_test_split
X = preProc_data.iloc[:,:6].values
y = preProc_data.iloc[:,6:7].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
# X_train.to_csv("X_train.csv")
# X_test.to_csv("X_test.csv")
# y_train.to_csv("y_train.csv")
# y_test.to_csv("y_test")
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=600,random_state=0)
rfc.fit(X_train,y_train)
predictions = rfc.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,predictions)
print('accuracy_RandForrest: %.3f' %accuracy)
```

```python
plt.title('Comparison of Y values in test and the Predicted values')
plt.ylabel('Test Set')
plt.xlabel('Predicted values')
plt.plot(y_test, '.',predictions , 'x')
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train,y_train)
predictions = decision_tree.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,predictions)
print('accuracy_DecisionTree: %.3f' %accuracy)
```

```python
plt.title('Comparison of Y values in test and the Predicted values')
plt.ylabel('Test Set')
plt.xlabel('Predicted values')
plt.plot(y_test, '.',predictions , 'x')
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
